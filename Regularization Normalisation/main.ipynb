```python
import torch
from src.dataLoader import Loader
from src.plots import Plots
from src.model import Model_loader
from src.optimise import learner

import warnings
warnings.filterwarnings('ignore')
```

```python
train_transform,test_transform = Loader.transform()
train_data,test_data = Loader.Loader(train_transform,test_transform)
```

```python
Plots.sampleVisual(train_data)
```

```python
use_cuda = torch.cuda.is_available()
device = 'cuda:0' if use_cuda else 'cpu'
```

```python
model = Model_loader.models("LN",device)
(ln_train_acc, ln_train_losses, ln_test_acc, ln_test_losses), ln_model= learner(model, train_data, test_data, 0, 20, device)
```

```python
model = Model_loader.models("GN",device)
(gn_train_acc, gn_train_losses, gn_test_acc, gn_test_losses), gn_model = learner(model, train_data, test_data, 0, 20, device)
```

```python
model = Model_loader.models("BN",device)
(bn_train_acc, bn_train_losses, bn_test_acc, bn_test_losses), bn_model = learner(model, train_data, test_data, 0.001, 20, device)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(18,5))
ax1 = ax[0]
ax1.set_title("Loss")
ax1.plot(ln_test_losses, color="blue", label="LN Test Loss")
ax1.plot(gn_test_losses, color="lime", label="GN Test Loss")
ax1.plot(bn_test_losses, color="red", label="BN Test Loss")
ax1.legend()
ax2 = ax[1]
ax2.set_title("Accuracy")
ax2.plot(ln_test_acc, color="red", label="LN Test Accuracy")
ax2.plot(gn_test_acc, color="crimson", label="GN Test Accuracy")
ax2.plot(bn_test_acc, color="lime", label="BN Test Accuracy")
ax2.legend()

plt.show()
```
